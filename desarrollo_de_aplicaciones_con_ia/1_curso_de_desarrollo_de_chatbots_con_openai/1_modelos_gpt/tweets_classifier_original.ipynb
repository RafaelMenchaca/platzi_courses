## Clasificador de Tweets con Completion API (Legacy)

```python
response = openai.completions.create(
  model="gpt-3.5-turbo-instruct",
  prompt="Decide si el sentimiento de un Tweet es positivo, neutral, o negativo. \
  \n\nTweet: \"Esto esta muy mal y no deberian poner asi los cursos de ia no me gusta nada 1 estrella.\
  \"\nSentiment:",
  temperature=0,
  max_tokens=60,
  top_p=1.0,
  frequency_penalty=0.5,
  presence_penalty=0.0
)
```
